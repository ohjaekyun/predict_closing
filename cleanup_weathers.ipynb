```python
import numpy as np
import pandas as pd
```

```python
df_weather = pd.read_csv('data/monthly_weather.csv', index_col=0)
df_weather.head()
```

```python
df_weather.tail()
```

```python
print(len(df_weather))
df_weather.isnull().sum()
```

```python
df_weather = df_weather.drop(columns=['stnko'])
```

```python
arr_isnull = df_weather.isnull().sum()
arr_isnull[arr_isnull != 0]
```

```python
def weather_at(dataframe, year, month):
    df = dataframe[dataframe['year'] == year]
    return df[df['month'] == month].reset_index(drop=True)
```

```python
df = weather_at(df_weather, 2016, 1)
df.head()
```

```python
df.mean().to_numpy()
```

```python
ccdf = pd.DataFrame([df.mean()], columns=df.columns)
ccdf.head()
```

```python
ccdf['year'] = 2015
ccdf.head()
```

```python
total_rows = []
years = np.arange(2015, 2023)
months = np.arange(1, 13)
for year in years:
    if year == 2022:
        cur_months = np.arange(1, 10)
    else:
        cur_months = months
    for month in cur_months:
        cur_df = weather_at(df_weather, year, month)
        if len(cur_df) == 0:
            cur_df1 = df_weather[df_weather['month'] == month]
            arr_for_cur_row = cur_df1.mean().to_numpy()
            arr_recent = total_rows[-1].to_numpy()
            recent_month = total_rows[-1].month.item()
            arr_recent_compare = df_weather[df_weather['month'] == recent_month].mean().to_numpy()
            arr_final = arr_for_cur_row + arr_recent - arr_recent_compare
            cur_row = pd.DataFrame(arr_final, columns=cur_df.columns)
        else:
            cur_row = pd.DataFrame([cur_df.mean()], columns=cur_df.columns)
        cur_row['year'] = year
        cur_row['month'] = month
        total_rows.append(cur_row)
```

```python
weather_at(df_weather, 2016, 3)
```

```python
total_rows
```

```python
total_df =  pd.concat(total_rows, ignore_index=True)
total_df.head()
```

```python
total_df.to_csv('data/average_weather.csv')
```
