```python
import pandas as pd
import numpy as np
```

```python
df_death = pd.read_csv('data/upper_death_rate_company.csv')
```

```python
df_death.head(10)
```

```python
df_death.columns
```

```python
cmp_death = df_death['CMP_NM']
```

```python
arr_cmp_death = cmp_death.to_numpy()
```

```python
all_list_cmps = []
for cmp in arr_cmp_death:
    if '(원청)' in cmp:
        all_list_cmps.append(cmp.split('(원청)'))
    else:
        all_list_cmps.append([cmp])
```

```python
all_list_cmps
```

```python
sfxs =['㈜', '(주)']
```

```python
columns = list(df_death.columns)
columns.append('IS_UNDER')
new_df_death = pd.DataFrame(columns=columns)
```

```python
def get_clean_cmp_nm(cmp_name):
    for sfx in sfxs:
        cmp_name = cmp_name.replace(sfx, '')
    if '(' in cmp_name:
        splits = cmp_name.split('(')
        first_part = splits[0]
        if len(first_part) == 0:
            cmp_name = splits[1].split(')')[1]
        else:
            cmp_name = first_part
    return cmp_name
```

```python
for i, cmps in enumerate(all_list_cmps):
    if len(cmps) == 1:
        original_row = pd.Series(df_death.iloc[i])
        cmp_name = cmps[0]
        if '(하청)' in cmp_name:
            cmp_name = cmp_name.split('(하청)')[0]
            is_under = True
        else:
            is_under = False
        cmp_name = get_clean_cmp_nm(cmp_name)
        row = pd.DataFrame(np.zeros([1, len(columns)]), columns=columns)
        row['year'] = original_row['year']
        row['num_death'] = original_row['num_death']
        row['death_rate_bp'] = original_row['death_rate_bp']
        row['average_death_rate_bp_by_size'] = original_row['average_death_rate_bp_by_size']
        row['IS_UNDER'] = is_under
        row['CMP_NM'] = cmp_name
        new_df_death = pd.concat([new_df_death, row], ignore_index=True)
    else:
        j = 0
        original_row = pd.Series(df_death.iloc[i])
        for cmp in cmps:
            cmp_name = cmp.lstrip(',')
            cmp_name = cmp_name.lstrip(' ')
            cmp_name = cmp_name.lstrip(',')
            if '(하청)' in cmp_name:
                cmp_name = cmp_name.split('(하청)')[0]
                is_under = True
            else:
                is_under = False
            if len(cmp_name) == 0:
                continue
            if cmp_name[0] == '(' and cmp_name[-1] == ')':
                continue
            parts = cmp_name.split(',')
            for part in parts:
                if len(part) > 1:
                    part = get_clean_cmp_nm(part)
                    if len(part) > 2:
                        original_row = pd.Series(df_death.iloc[i])
                        row = pd.DataFrame(np.zeros([1, len(columns)]), columns=columns)
                        row['year'] = original_row['year']
                        row['num_death'] = original_row['num_death']
                        row['death_rate_bp'] = original_row['death_rate_bp']
                        row['average_death_rate_bp_by_size'] = original_row['average_death_rate_bp_by_size']
                        row['IS_UNDER'] = is_under
                        row['CMP_NM'] = part
                        new_df_death = pd.concat([new_df_death, row], ignore_index=True)
```

```python
new_df_death.to_csv('data/high_death_rate_company.csv')
```
