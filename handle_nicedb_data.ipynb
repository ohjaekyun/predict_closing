```python
import pandas as pd
```

```python
data_out_biz = pd.read_excel('data/액티브 중소법인 재무보유.xlsx')
data_all_biz = pd.read_excel('data/휴폐업 중소법인 재무보유.xlsx', sheet_name='기업개요')
data_biz_history = pd.read_excel('data/액티브 중소법인 재무보유.xlsx', sheet_name='휴폐업이력')
```

```python
bizno_is_out = data_out_biz.BIZ_NO.unique()
```

```python
data_all_biz['IS_LP'] = (data_all_biz.ESTB_GB == 2).to_numpy() 
data_all_biz['IS_LC'] = (data_all_biz.ESTB_GB == 4).to_numpy() 
data_all_biz['IS_LTD'] = (data_all_biz.ESTB_GB == 1).to_numpy()
data_all_biz['IS_SP'] = (data_all_biz.ESTB_GB == 5).to_numpy() 
data_all_biz['IS_ETC'] = (data_all_biz.ESTB_GB == 99).to_numpy() 
```

```python
(((data_all_biz.CMP_SFIX_NM == '(자)').to_numpy() + (data_all_biz.CMP_SFIX_NM == '(합)').to_numpy() 
+ (data_all_biz.CMP_SFIX_NM == '(유)').to_numpy() + (data_all_biz.CMP_SFIX_NM == '(주)').to_numpy() 
+ (data_all_biz.CMP_PFIX_NM == '(복)').to_numpy()) >=2).sum()
```

```python
cns = []
ceonames = data_all_biz.CEO_NM.to_numpy()
for ceo in ceonames:
    if type(ceo) == str:
        try:
            cn = ceo[-2]
            cns.append(int(cn) + 1)
        except:
            cns.append(1)
    else:
        cns.append(None)
cns
```

```python
nhp=data_all_biz.HOMEPAGE_URL.to_numpy()
has_hp = []
for hp in nhp:
    if type(hp) == str:
        if len(hp) > 3:
            if hp[-1] == '/':
                has_hp.append(True)
                continue
            else:
                try:
                    if hp[-1].lower() in 'abcdefghijklmnopqrstuvwxyz':
                        has_hp.append(True)
                        continue
                except:
                    pass
    has_hp.append(False)
has_hp
```

```python
data_all_biz['HAS_HOMEPAGE'] = has_hp
data_all_biz['CEO_CNT'] = cns
```

```python
out_con_nums = data_out_biz.BIZ_NO.to_numpy()
```
