```python
df_opened_summary = pd.read_csv('data/opened_company_summary_final.csv', index_col=0)
df_opened_summary.head()
```

```python
df_total_summary = pd.concat([df_opened_summary, df_closed_summary]).reset_index(drop=True)
df_total_summary.head(10)
```

```python
df_fin = pd.read_csv('data/finantial_data_final.csv', index_col=0)
df_fin
```

```python
merged_df1 = df_fin.merge(df_histories, how='right', left_on=['BIZ_NO', 'year'], right_on=['BIZ_NO', 'year'])
merged_df1.head(7)
```

```python
merged_df1.label.sum()
```

```python
na_list = []
vals = []
for idx, val in zip(merged_df1.isna().sum().index, merged_df1.isna().sum().values):
    if val > 0:
        na_list.append(idx)
        vals.append(val)
```
