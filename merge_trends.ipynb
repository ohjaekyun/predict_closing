```python
df_google = pd.read_csv('./data/google_trends.csv', index_col=0)
df_naver = pd.read_csv('./data/naver_trends.csv', index_col=0)
```

```python
df_google.head()
```

```python
df_google1 = df_google[df_google['year'] > 2015].reset_index(drop=True, inplace=False)
df_google1.head()
```

```python
df_naver.head()
```

```python
df_naver1 = df_naver.drop(columns=['year', 'month'])
naver_columns = list(df_naver1.columns)
naver_columns
```

```python
naver_columns1 = []
for col_name in naver_columns:
    naver_columns1.append('naver_' + col_name)
naver_columns1
```

```python
df_naver1.columns = naver_columns1
df_naver1.head()
```

```python
total_trends = pd.concat([df_google1, df_naver1], axis=1)
total_trends.head()
```
