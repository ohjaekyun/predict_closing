```python
drop_columns = []
for col, value in zip(df_fin.isna().sum().index, df_fin.isna().sum().values):
    if value <= 10000:
        df_fin[col] = df_fin[col].fillna(df_fin[col].mean())
    else:
        drop_columns.append(col)
df_fin1 = df_fin.drop(columns=drop_columns)
df_fin1.head(10)
```

```python
df_fin1['year'] = df_fin1.STT_YR_MO.map(lambda x: int(str(x)[:4]))
```

```python
df_fin1 = df_fin1.drop(columns=['STT_YR_MO'])
df_fin1.head(10)
```

```python
df_fin1.to_csv('data/finantial_data_final.csv')
```
