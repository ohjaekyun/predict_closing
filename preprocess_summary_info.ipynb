```python
na_columns = []
etc_na_columns = []
for nm, value in zip(series_na.index, series_na.values):
    if value > 1000:
        na_columns.append(nm)
    elif value > 0:
        etc_na_columns.append(nm)
```

```python
etc_na_columns
```

```python
non_distinct_columns = []
for uq, value in zip(df_closed.nunique().index, df_closed.nunique().values):
    if value == 1:
        if uq != 'label':
            non_distinct_columns.append(uq)
print(non_distinct_columns)
```

```python
etc_na_columns1 = []
for col in etc_na_columns:
    if col not in ['LIST_CD', 'IND_CD2', 'IND_CD3', 'HDOF_CMP_CD', 'LIST_DATE', 'NATN_NM', 'MN_BIZ_CONT', 'ESTB_GB', 'CMP_SCL', 'PBCO_GB', 'PSN_CORP_GB', 'IND_CD_ORDR', 'PB_ORG_TYP']:
        etc_na_columns1.append(col)
print(etc_na_columns1)
```

```python
df_closed1 = df_closed.drop(columns=['LIST_CD', 'IND_CD2', 'IND_CD3', 'HDOF_CMP_CD', 'LIST_DATE', 'NATN_NM', 'MN_BIZ_CONT', 'ESTB_GB', 'CMP_SCL', 'PBCO_GB', 'PSN_CORP_GB', 'IND_CD_ORDR', 'PB_ORG_TYP'])
```

```python
df_closed1 = df_closed1.replace({
    'BZ_TYP':{
        'A': 0,
        'B': 1,
        'C': 2,
        'F': 3,
        'I': 4,
        'M': 5,
        'O': 6,
        'S': 7,
        'V': 8,
        }
    })
```

```python
for col in ['FR_IVST_CORP_YN', 'VENT_YN', 'MDSCO_PRTC_YN']:
    df_closed1 = df_closed1.replace({col:{'N': 0, 'Y': 1}})
```

```python
for col in ['IS_LP', 'IS_LC', 'IS_LTD', 'IS_SP', 'IS_ETC', 'HAS_HOMEPAGE']:
    df_closed1 = df_closed1.replace({col:{False: 0, True: 1}})
```

```python
series_na = df_opened.isna().sum()
print(series_na.index)
```

```python
df_opened1 = df_opened.drop(columns=['LIST_CD', 'IND_CD2', 'IND_CD3', 'HDOF_CMP_CD', 'LIST_DATE', 'NATN_NM', 'MN_BIZ_CONT', 'ESTB_GB', 'CMP_SCL', 'PBCO_GB', 'PSN_CORP_GB', 'IND_CD_ORDR', 'PB_ORG_TYP'])
len(df_opened1.columns)
```

```python
df_opened1.iloc[:10,:10].head(10)
```

```python
df_opened1 = df_opened1.replace({
    'BZ_TYP':{
        'A': 0,
        'B': 1,
        'C': 2,
        'F': 3,
        'I': 4,
        'M': 5,
        'O': 6,
        'S': 7,
        'V': 8,
        }
    })
for col in ['FR_IVST_CORP_YN', 'VENT_YN', 'MDSCO_PRTC_YN']:
    df_opened1 = df_opened1.replace({col: {'N': 0, 'Y': 1}})
for col in ['IS_LP', 'IS_LC', 'IS_LTD', 'IS_SP', 'IS_ETC', 'HAS_HOMEPAGE']:
    df_opened1 = df_opened1.replace({col: {False: 0, True: 1}})
```

```python
df_opened1.iloc[:10,:10].head(10)
```

```python
df_opened1.iloc[:10,10:].head(10)
```

```python
for col in etc_na_columns1:
    meanvalue = (df_closed1[col].mean() * len(df_closed1) + df_opened1[col].mean() * len(df_opened1)) / (len(df_closed1) + len(df_opened1))
    df_closed1[col] = df_closed1[col].fillna(meanvalue)
    df_opened1[col] = df_opened1[col].fillna(meanvalue)
```

```python
df_opened1.to_csv('data/opened_company_summary_final.csv')
df_closed1.to_csv('data/closed_company_summary_final.csv')
```
